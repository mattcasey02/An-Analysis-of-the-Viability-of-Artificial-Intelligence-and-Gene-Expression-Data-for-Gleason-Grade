```python
import numpy as np
import pandas as pd
#pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
```

```python
data = pd.read_csv('../Data/nationwidechildrens.org_clinical_patient_prad.txt' ,delimiter='\t')
TCGAHealthy = pd.read_pickle('../Data/TCGAHealthy.pkl')
TCGACancer = pd.read_pickle('../Data/TCGACancer.pkl')
GTEX = pd.read_pickle('../Data/GTEX.pkl')
```

```python
data
```

```python
gleason = data[['gleason_pattern_primary','gleason_pattern_secondary','bcr_patient_barcode']]
```

```python
gleason
```

```python
gleason.drop([0,1], axis=0, inplace=True)
```

```python
gleason.set_index('bcr_patient_barcode', inplace=True)
```

```python
gleason['grade'] = gleason['gleason_pattern_primary'] + "-" + gleason['gleason_pattern_secondary']
```

```python
gleason['grade'].value_counts()
```

```python
gleason.drop(['gleason_pattern_primary','gleason_pattern_secondary'], axis=1,inplace=True)
```

```python
gleason['Target'] = gleason['grade'].apply(lambda x: 1 if (x=='2-4') 
                                           else(1 if (x=='3-3') 
                                                else(2 if (x=='3-4') 
                                                     else(3 if (x=='4-3') 
                                                          else(4 if(x=='5-3') 
                                                               else(4 if (x=='3-5') 
                                                                    else(4 if (x=='4-4') 
                                                                         else(5 if (x=='5-5') 
                                                                              else(5 if (x=='5-4') 
                                                                                   else(5 if(x=='4-5') else 6))))))))))
```

```python
gleason['Target'].value_counts()
```

```python
gleason.drop('grade', axis=1, inplace=True)
```

```python
healthy = GTEX.append(TCGAHealthy)
```

```python
healthy
```

```python
TCGACancer
```

```python
gleason['bcr_patient_barcode'] = gleason.index
```

```python
TCGACancer['bcr_patient_barcode'] = TCGACancer.index
```

```python
TCGACancer['bcr_patient_barcode'] = TCGACancer['bcr_patient_barcode'].apply(lambda x:x[0:12])
```

```python
cancer = pd.merge(TCGACancer, gleason, on='bcr_patient_barcode', how='inner')
```

```python
cancer
```

```python
cancer.drop('Target_x', axis = 1, inplace=True)
```

```python
cancer['Target'] = cancer['Target_y']
```

```python
cancer.drop('Target_y', axis = 1, inplace=True)
```

```python
cancer.drop('bcr_patient_barcode', axis = 1, inplace=True)
```

```python
cancer
```

```python
cancer.to_pickle('../Data/CancerMulti.pkl')
healthy.to_pickle('../Data/HealthyMulti.pkl')
```
